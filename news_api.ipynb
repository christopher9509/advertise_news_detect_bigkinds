```python
json_file = response.json()
json_file
```

```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#   from": "2022-08-11"- "until": "2022-09-20"
df = json_normalize(json_file['return_object']['documents'])
df = df.drop(labels='category_incident',axis=1)
df1 = df
```

```python
#   "from": "2022-07-11", "until": "2022-08-12"
df = json_normalize(json_file['return_object']['documents'])
df = df.drop(labels='category_incident',axis=1)
df2 = df
```

```python
#   "from": "2022-06-11", "until": "2022-07-12"
df = json_normalize(json_file['return_object']['documents'])
df = df.drop(labels='category_incident',axis=1)
df3 = df
```

```python
#   "from": "2022-05-11", "until": "2022-06-12"
df = json_normalize(json_file['return_object']['documents'])
df = df.drop(labels='category_incident',axis=1)
df4 = df
```

```python
news_data = pd.concat([df1, df2, df3, df4], axis = 0)
news_data
```

```python
news_data = news_data.drop(labels='news_id',axis=1)
news_data = news_data.drop(labels='enveloped_at',axis=1)
news_data = news_data.drop(labels='provider_news_id',axis=1)
news_data = news_data.drop(labels='published_at',axis=1)
```

```python
news_data = news_data.sort_values('dateline', ascending=False).reset_index()
news_data = news_data.drop(labels='index',axis=1)
```
